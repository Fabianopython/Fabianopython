```python
# abre a conexao com o banco de dados
import sqlite3
import pandas as pd
conn = sqlite3.connect( 'database.db' )

# consulta dos dados nos banco de dados
consulta_atividade = """
SELECT
fa.loyalty_number,
flh.loyalty_card,
flh.gender,
fa.points_accumulated
FROM
flight_activity fa LEFT JOIN flight_loyalty_history flh
ON (fa.loyalty_number = flh.loyalty_number)
WHERE
flh.salary > 132000
AND flh.marital_status = "Married"
AND flh.education = "Master"
AND ( fa.total_flights = fa.flights_booked )
"""
# execucao da consulta
df_atividade = pd.read_sql_query( consulta_atividade, conn )
# fecha a conexao com o banco de dados
conn.close()
# confere se deu certo
df_atividade.head()
```

```python
# Importando as bibliotecas
import sqlite3
import pandas as pd

# Abrindo a conexão com o banco de dados
conn = sqlite3.connect("database.db")

# Coletando os dados
consulta_atividade = """
SELECT *
FROM flight_activity fa
LEFT JOIN flight_loyalty_history flh
ON fa.loyalty_number = flh.loyalty_number
"""
# Executando a consulta
df_atividade = pd.read_sql_query(consulta_atividade, conn)

# Selecionando colunas de uma planilha
colunas = ['year', 'month', 'flights_booked',
           'flights_with_companions', 'total_flights',
           'distance', 'points_accumulated',
           'points_redeemed', 'dollar_cost_points_redeemed',
           'salary', 'clv', 'enrollment_year',
           'enrollment_month', 'loyalty_card']

# Selecionando linhas de uma planilha
linhas = ~df_atividade['salary'].isna()

# Aplicando a seleção de linhas e colunas
df_dados_limpos = df_atividade.loc[linhas, colunas]

# Verificando o número de linhas vazias
print(df_dados_limpos.isna().sum())
```
